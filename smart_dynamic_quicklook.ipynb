```python
import smart
from smart import plot_complete_smart_spectra_interactive, plot_smart_data_interactive
from functions_jr import make_dir
import os
import numpy as np
import pandas as pd
import holoviews as hv
from holoviews import opts
from typing import Union
```

```python
raw_path, pixel_path, calib_path, data_path, plot_path = smart.set_paths()
flight = "flight_00"
ql_path = f"{plot_path}/quicklooks/{flight}"
make_dir(ql_path)
calibrated_path = smart.get_path("calibrated")
inpath = f"{calibrated_path}/{flight}"
spectra_files = [file for file in os.listdir(inpath) if "VNIR" in file]
```
